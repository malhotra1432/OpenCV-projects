```python
import cv2
```

```python
img1 = cv2.imread("./images/img.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("./images/eagle_1.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
# Blending images of same size
img1 = cv2.resize(img1,(1200,1200))
img2 = cv2.resize(img2,(1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blended_img = cv2.addWeighted(src1=img1,alpha=0.6,src2=img2,beta=0.4,gamma=0)
```

```python
plt.imshow(blended_img)
```
